### Calling Unique Categorical Values

```python
#calling unique categorical values of the descriptive dataset
for column in ds_cust_des.columns:
    if ds_cust_des[column].dtype==object:
        print(str(column)+":"+str(ds_cust_des[column].unique()))
        print(ds_cust_des[column].value_counts())
        print("************************************************")
        print("\n")
```
